```python
import pandas as pd
from sklearn.impute import SimpleImputer
data=pd.read_excel("C://Users//Siri//Downloads//Strikers_performance.xlsx")
data.head()
```

```python
#DATA CLEANING

# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

# Separate numeric and nominal columns
numeric_columns = data.select_dtypes(include=['int64', 'float64']).columns
nominal_columns = data.select_dtypes(include=['object']).columns

# Handle missing values
numeric_imputer = SimpleImputer(strategy='median')
nominal_imputer = SimpleImputer(strategy='most_frequent')

# Apply imputers
data[numeric_columns] = numeric_imputer.fit_transform(data[numeric_columns])
data[nominal_columns] = nominal_imputer.fit_transform(data[nominal_columns])

# Verify if missing values are handled
print("Missing Values After Imputation:\n", data.isnull().sum())
```

```python
# Check the current data types of all columns
print("Data Types Before Correction:\n", data.dtypes)

# List of specific columns to convert to integers
columns_to_convert = [
    'Goals Scored', 'Assists', 'Shots on Target', 'Movement off the Ball',
    'Hold-up Play', 'Aerial Duels Won', 'Defensive Contribution',
    'Big Game Performance', 'Impact on Team Performance', 'Off-field Conduct'
]

# Convert the listed columns to integer type, if present in the dataset
for column in columns_to_convert:
    if column in data.columns:
        data[column] = data[column].astype(int)

# Verify the data type changes
print("Data Types After Correction:\n", data.dtypes)
```

```python
# DESCRIPTIVE STATISTICS

# Perform descriptive analysis and round output to 2 decimal points
descriptive_stats = data.describe().round(2)

# Display the descriptive statistics
print("Descriptive Statistics:\n", descriptive_stats)
```

```python
# DATA VISUALIZATION 

import matplotlib.pyplot as plt

# Percentage analysis for Footedness
footedness_percentage = data['Footedness'].value_counts(normalize=True) * 100

# Pie chart for Footedness
plt.figure(figsize=(6, 6))
footedness_percentage.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Percentage Distribution of Footedness')
plt.ylabel('')  # Remove default ylabel
plt.show()
```

```python
import seaborn as sns

# Countplot for Footedness distribution by Nationality
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Nationality', hue='Footedness', palette='viridis')
plt.title('Distribution of Players\' Footedness Across Nationalities')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.legend(title='Footedness')
plt.show()
```

```python
# STATISTICAL ANALYSIS

# 1.Determine which nationality strikers have the highest average number of goals scored
# Grouping by Nationality and calculating average goals scored
avg_goals_by_nationality = data.groupby('Nationality')['Goals Scored'].mean().round(2)

# Finding the nationality with the highest average goals scored
highest_avg_goals = avg_goals_by_nationality.idxmax()
highest_avg_goals_value = avg_goals_by_nationality.max()

print(f"Nationality with the highest average goals scored: {highest_avg_goals} ({highest_avg_goals_value} goals)")
```

```python
#2. Calculate the average conversion rate for players based on their footedness

# Grouping by Footedness and calculating the average conversion rate
avg_conversion_rate = data.groupby('Footedness')['Conversion Rate'].mean().round(2)

print("Average Conversion Rate by Footedness:")
print(avg_conversion_rate)
```

```python
# 3. Find whether there is a significant difference in consistency rates among strikers from various nationalities
#Steps:
#Check Normality (Shapiro-Wilk Test):
from scipy.stats import shapiro

# Shapiro-Wilk test for consistency rates
stat, p_value = shapiro(data['Consistency'])
print(f"Shapiro-Wilk Test p-value: {p_value}")
if p_value > 0.05:
    print("Data is normally distributed.")
else:
    print("Data is not normally distributed.")
    
#Check Homogeneity of Variances (Levene’s Test):

from scipy.stats import levene

# Levene's test for homogeneity of variances
stat, p_value = levene(*[group['Consistency'].values for _, group in data.groupby('Nationality')])
print(f"Levene’s Test p-value: {p_value}")
if p_value > 0.05:
    print("Homogeneity of variances is assumed.")
else:
    print("Homogeneity of variances is not assumed.")
    
# Perform ANOVA (or Welch’s ANOVA if variances are unequal):

from scipy.stats import f_oneway

# ANOVA test
anova_result = f_oneway(*[group['Consistency'].values for _, group in data.groupby('Nationality')])
print(f"ANOVA p-value: {anova_result.pvalue}")
if anova_result.pvalue < 0.05:
    print("There is a significant difference in consistency rates among nationalities.")
else:
    print("No significant difference in consistency rates among nationalities.")
```

```python
# 4. Check for significant correlation between Hold-up Play and Consistency Rate
# Steps:
#Check Linearity: Plot a scatterplot.
sns.scatterplot(data=data, x='Hold-up Play', y='Consistency', color='blue')
plt.title("Scatterplot: Hold-up Play vs Consistency")
plt.show()

# Correlation Test:

from scipy.stats import pearsonr

# Pearson correlation
corr_coef, p_value = pearsonr(data['Hold-up Play'], data['Consistency'])
print(f"Correlation Coefficient: {corr_coef:.2f}, p-value: {p_value}")
if p_value < 0.05:
    print("Significant correlation between Hold-up Play and Consistency.")
else:
    print("No significant correlation between Hold-up Play and Consistency.")
```

```python
# 5. Check if Hold-up Play significantly influences Consistency Rate
#Steps:
# Fit a Linear Regression Model:

import statsmodels.api as sm

# Linear regression: Hold-up Play as predictor for Consistency
X = sm.add_constant(data['Hold-up Play'])  # Add constant term
y = data['Consistency']
model = sm.OLS(y, X).fit()

# Display regression results
print(model.summary())
```

```python
# FEATURE ENGINEERING

# 1. Create a new feature: Total Contribution Score

# Creating Total Contribution Score by summing up specific columns
data['Total Contribution Score'] = (
    data['Goals Scored'] +
    data['Assists'] +
    data['Shots on Target'] +
    data['Dribbling Success'] +
    data['Aerial Duels Won'] +
    data['Defensive Contribution'] +
    data['Big Game Performance'] +
    data['Consistency']
)
print("New feature 'Total Contribution Score' added successfully.")

# 2. Encode Footedness and Marital Status using LabelEncoder

from sklearn.preprocessing import LabelEncoder

# Encoding Footedness
label_encoder = LabelEncoder()
data['Footedness_Encoded'] = label_encoder.fit_transform(data['Footedness'])

# Encoding Marital Status
data['Marital_Status_Encoded'] = label_encoder.fit_transform(data['Marital Status'])

print("Footedness and Marital Status encoded successfully.")

# 3. Create dummy variables for Nationality

# Creating dummy variables for Nationality
data = pd.get_dummies(data, columns=['Nationality'], prefix='Nationality', drop_first=True)

print("Dummy variables for Nationality created and added to the dataset.")
```

```python
# CLUSTERING ANALYSIS 

#1.Select Features and Drop Striker_ID
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Dropping unnecessary column
clustering_features = data.drop(['Striker_ID'], axis=1)

# Ensure only numerical columns are included for clustering
clustering_features = clustering_features.select_dtypes(include=['float64', 'int64'])

print("Features for clustering selected.")

#2. Calculate WCSS and Visualize the Elbow Chart
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(clustering_features)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Chart
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Chart for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid()
plt.show()

# From the chart, identify the "elbow point" where the WCSS significantly decreases and selecting n_clusters=2.

# Build KMeans Clustering and Add Labels

# Optimal number of clusters
kmeans = KMeans(n_clusters=2, random_state=42)
data['Cluster'] = kmeans.fit_predict(clustering_features)

# Calculate Average Total Contribution Score for Each Cluster
average_contribution = data.groupby('Cluster')['Total Contribution Score'].mean()
print("Average Total Contribution Score by Cluster:\n", average_contribution)

#Assign Striker Types and Drop Clusters

# Assign striker types
data['Striker_Type'] = data['Cluster'].map({0: 'Best strikers', 1: 'Regular strikers'})

# Drop the Cluster column
data = data.drop(['Cluster'], axis=1)

# Encode Striker_Type for modeling
data['Striker_Type_Encoded'] = data['Striker_Type'].map({'Best strikers': 1, 'Regular strikers': 0})

print("Striker types assigned and new column added.")
data.head()
```

```python
# MACHINE LEARNING MODEL


# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt


# Step 1: Select features (X) and target (y), and drop unnecessary columns
X = data.drop(['Striker_ID', 'Striker_Type', 'Footedness', 'Marital Status'], axis=1)  # Drop unnecessary columns
y = data['Striker_Type_Encoded']  # Use encoded target column

# Step 2: Perform feature scaling with StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Split the data into train and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Step 4: Build the logistic regression model
log_reg = LogisticRegression(random_state=42)
log_reg.fit(X_train, y_train)

# Step 5: Make predictions on the test data
y_pred = log_reg.predict(X_test)

# Step 6: Evaluate the model by calculating the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the Logistic Regression Model: {accuracy * 100:.2f}%")

# Step 7: Create the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Step 8: Visualize the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=['Regular strikers', 'Best strikers'])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix for Logistic Regression Model")
plt.show()


```
